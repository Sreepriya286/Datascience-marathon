```python
import os
import numpy as np
import cv2

classes = []

def load_datas(train_path, test_path):
    train_images, train_labels = [], []
    test_images = []

    # Load training data
    for class_name in os.listdir(train_path):
        class_path = os.path.join(train_path, class_name)

        # Ensure it's a directory for class
        if os.path.isdir(class_path):
            classes.append(class_name)  # Append class names

            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)

                # Read image
                image = cv2.imread(img_path)
                if image is not None:  # Check if the image was loaded
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
                    image = cv2.resize(image, (100, 100))  # Resize to 100x100
                    train_images.append(image)
                    train_labels.append(class_name)

    # Load test data
    for img_name in os.listdir(test_path):
        img_path = os.path.join(test_path, img_name)

        # Read image
        image = cv2.imread(img_path)
        if image is not None:  # Check if the image was loaded
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
            image = cv2.resize(image, (100, 100))  # Resize to 100x100
            test_images.append(image)

    return np.array(train_images) / 255.0, np.array(train_labels), np.array(test_images) / 255.0
```

```python
train_path = '/content/train/train'  # Path to the outer train folder
test_path = '/content/test/test'     # Path to the outer test folder
```

```python
train_images,train_labels,test_images = load_datas(train_path,test_path)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

def plot_counts(labels):
    plt.figure(figsize=(12, 6))

    # Create a count plot
    sns.countplot(y=labels, order=sorted(set(labels)))
    plt.title('Fruit Counts')
    plt.ylabel('Fruits', fontsize=15)
    plt.xlabel('Count', fontsize=15)
    plt.tight_layout()

    # Create a DataFrame for the counts
    count_data = pd.Series(labels).value_counts().reset_index()
    count_data.columns = ['Fruit', 'Count']

    # Adjust space between the two plots
    plt.subplots_adjust(bottom=0.2)  # Add space below the count plot

    # Display the count data as a table using matplotlib
    plt.figure(figsize=(12, 6))
    table = plt.table(cellText=count_data.values, colLabels=count_data.columns, cellLoc='center', loc='center')
    table.auto_set_font_size(False)  # Allow manual font size
    table.set_fontsize(12)  # Set the font size for the table
    plt.axis('off')  # Turn off the axis

    # Center the title
    plt.title('Fruit Counts Table', fontsize=15, loc='center', pad=50)  # Add padding for better positioning
    plt.show()

# Example usage
plot_counts(train_labels)
```

```python
import random
import matplotlib.pyplot as plt

def show_one_random_image_per_class(images, labels):
    class_to_images = {}

    # Group images by class
    for img, label in zip(images, labels):
        if label not in class_to_images:
            class_to_images[label] = []
        class_to_images[label].append(img)

    # Get the number of classes and determine the layout for the subplots
    num_classes = len(class_to_images)

    # Set up the plot with a dynamic grid size based on the number of classes
    cols = 5
    rows = (num_classes + cols - 1) // cols  # Ceiling division for rows

    plt.figure(figsize=(15, rows * 3))  # Adjust height based on number of rows

    # Select one random image from each class and display it
    for i, (label, imgs) in enumerate(class_to_images.items()):
        random_index = random.randint(0, len(imgs) - 1)  # Random index for images of this class
        plt.subplot(rows, cols, i + 1)
        plt.imshow(imgs[random_index])
        plt.title(label)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage:
show_one_random_image_per_class(train_images, train_labels)
```

```python
# Create the dictionary to map class names to integer labels
class_labels = {cname: i for i, cname in enumerate(classes)}

# Encode the train labels using this dictionary
train_labels = [class_labels[label] for label in train_labels]

# Convert to numpy array for compatibility with model training
train_labels = np.array(train_labels)

# Print the unique encoded labels to check the mapping
print(f"Encoded train labels: {np.unique(train_labels)}")


```

```python
from sklearn.model_selection import train_test_split

# Assuming train_images and train_labels are your training dataset
X_train, X_val, y_train, y_val = train_test_split(
    train_images,
    train_labels,
    test_size=0.2,  # Use 20% of data for validation
    stratify=train_labels,  # Maintain class distribution
    random_state=42  # For reproducibility
)

# Check the shapes of the splits
print(f'Training images shape: {X_train.shape}, Training labels shape: {y_train.shape}')
print(f'Validation images shape: {X_val.shape}, Validation labels shape: {y_val.shape}')
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def create_cnn_model(input_shape, num_classes):
    model = models.Sequential()

    # Convolutional Layer 1
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))

    # Convolutional Layer 2
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Convolutional Layer 3
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten the output
    model.add(layers.Flatten())

    # Fully Connected Layer
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.5))  # Dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax'))  # Output layer

    return model

# Define the model
input_shape = (100, 100, 3)  # Adjust based on your image size
num_classes = len(set(train_labels))  # Number of unique classes
cnn_model = create_cnn_model(input_shape, num_classes)

# Compile the model
cnn_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',  # Use 'categorical_crossentropy' if using one-hot encoding
                  metrics=['accuracy'])

# Print the model summary
cnn_model.summary()
```

```python
# Train the model
history = cnn_model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=20,  # Adjust based on your needs
                        batch_size=16)  # Adjust based on your needs
```
